### Step 2. Introducing Vector layers

First, we will open some Vector layers and see how layer order is important for viewing data:

- Open **QGIS Desktop**, (not **Browser**)


- Click **Layer > Add Vector Layer > Browse**


- Navigate to **../Exercise1/data/Alaska.shp**


- Click **Open**


Up should come a plain map of Alaska; no bounderies or geographical features. 

- Repeat the above steps, this time opening the layers **/Alaska_counties.shp** and **/Alaska_Volcanoes.shp**


If you have a scroll mouse, you will notice the ablity to zoom in/out of the map projection with scolling the mouse. 

Lets look at the order of layers:

- We can click ** Control Rendering Order** underneath the ** Layer Order** Panel to allow access to the layers.


- If we Click and Drag the layers up or down within the box, we can change the order of the layers in the map projection. Try this with a few differents orders and notice how information is sometimes missing if the layers are not in the optimal order (ie. Some Volcanoes are missing). Other ways to add Vector files incude:

- Accessing files from the built-in **Broswer** situated in conjuction with the **Layer order** panel - this allows you to drag-and-drop layers into the projection panel. 


- Keyboard shortcut **Ctrl+Shift+V**


- On the layers toolbar you will see a symbol similar to a large **V** with a small **'plus sign'**. If you hover over each symbol their function descriptions will appear.


Also, you can also set uo your own panels, toolbars, and workspace by going Clicking on the top menu **View > Panels >** or **View > Toolbars >** and selecting or deselecting whichever ones you wish to have. This way you can always re-open a panel that may have been closed by mistake. 


#### Shapefiles:

So far the three layers we have imported have been 'shapefiles' (extension .shp). These are used to show point data (such as the volcanoes) or polygon data (such as the county boundaries). We will look later at raster data which is used for continous data such as areas and coverage of a phenomena.
