## Step 3: Producing an Intrepretable Map

We need to manipulate the projects layers to be able interep the data correctly, and to check if our results are the same as the USDM homepage.

Firstly, reorder the layers to ensure that point data is sitting above line data, which in turn is sitting above polygon data. This way no information is being masked by another layer.  

Secondly, we are going to change the stlye of the **Drought_Impacts** layer to make it easier to interpret. You may have noticed on the USDM site they provide a Drought.lyr file which provides they style and colour scheme for thier shapefiles. These appear to be ArcGIS compataible only, and will not be recognized by QGIS; but we will create our own that could then be saved for later use on other USDM files.


- Bring up the **Layer Properties** for **States_Boundaries** layer and change the Style from **Simple fill** to **Simple line**. Repeat the process for the **Alaska** layer. You can now reorder these above the Drought Monitor layer as these are effectivly line data layers now.


- In the **Layer Properties** for the Drought Monitor layer navigate to **Style** once again this time change the dropdown menu that is currently set on **Single Symbol** to **Graduated**.


- In the **Column** option select **DM**. These are the columns in the data file and DM represents the Drought Monitor level (0 to 5).


- From **Mode** select **Natural Breaks**. Double click on each of the **Label** rows to manually change the labels to 1, 2... 4 respectively. 


- From the **Color ramp** selections choose a sutible red / orange / brown graduated colour scheme. If you want to be really picky, you can select each value and specify the exact colour you wish. The image below shows a **Oranges** Color ramp with the Yellow and Red specified for D0 and D4 as per the USDM website. Any of the standard graduated colour scheme will help to present the information better than random colours or a divergent scheme.
